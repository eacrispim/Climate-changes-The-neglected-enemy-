```python
## calcular regressao para cada city
coefficients = []

for city in cities:
    x = merged_table[['YEAR']] # tem que ser um dataframe, por isso os dois colchetes [[]]
    y = merged_table[city] # tem que ser um pandas series, por isso 1 unico colchete

    model = LinearRegression()
    model.fit(X=x,y=y)

    coefficients.append(model.coef_[0])
    
list(zip(coefficients, cities))
```

```python
model = LinearRegression()

x = merged_table[['YEAR']] # tem que ser um dataframe, por isso os dois colchetes [[]]
y = merged_table['ANNUAL']

model.fit(x,y)

coefficients.append(model.coef_[0])
```
